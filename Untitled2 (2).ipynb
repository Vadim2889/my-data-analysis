```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ===== A. Загрузка =====
df = pd.read_csv("google_books_dataset.csv")

print(df.head())
print(df.tail())
print(df.info())

# ===== B. Обработка =====

# Пропуски
print(df.isna().sum())

# Числовые и категориальные столбцы
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include="object").columns

# Заполнение пропусков
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Дубликаты
print("Дубликаты:", df.duplicated().sum())
df = df.drop_duplicates()

# Выбросы (z-score ТОЛЬКО для чисел)
z = np.abs((df[num_cols] - df[num_cols].mean()) / df[num_cols].std())
df = df[(z < 3).all(axis=1)]

# Статистика
print(df.describe())
print(df.nunique())
print(df[num_cols].corr())

# ===== C. Визуализация (5 графиков) =====

# 1. Гистограмма
plt.hist(df[num_cols[0]])
plt.title("Histogram")
plt.show()

# 2. Линейный график
plt.plot(df[num_cols[0]].values)
plt.title("Line plot")
plt.show()

# 3. Столбчатая диаграмма
df[cat_cols[0]].value_counts().head(10).plot(kind="bar")
plt.title("Bar chart")
plt.show()

# 4. Boxplot
plt.boxplot(df[num_cols[0]])
plt.title("Boxplot")
plt.show()

# 5. Heatmap корреляций
plt.imshow(df[num_cols].corr())
plt.colorbar()
plt.title("Correlation heatmap")
plt.show()

# ===== D. Группировки =====

grouped = df.groupby(cat_cols[0])[num_cols[0]].agg(
    ["mean", "median", "count", "sum"]
)
print(grouped.head())

pivot = pd.pivot_table(
    df,
    values=num_cols[0],
    index=cat_cols[0],
    aggfunc=["mean", "count"]
)
print(pivot.head())
```
