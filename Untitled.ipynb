```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("google_books_dataset.csv")

print(df.head())
print(df.tail())
print(df.info())

print(df.isna().sum())

num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include="object").columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Дубликаты:", df.duplicated().sum())
df = df.drop_duplicates()

df = df[(df['list_price'] < 5000) & (df['average_rating'] > 0)]

print(df.describe())
print(df.nunique())
print(df[num_cols].corr())

plt.hist(df[num_cols[0]])
plt.title("Histogram")
plt.show()

plt.plot(df[num_cols[0]].values)
plt.title("Line plot")
plt.show()

df[cat_cols[0]].value_counts().head(10).plot(kind="bar")
plt.title("Bar chart")
plt.show()

plt.boxplot(df[num_cols[0]])
plt.title("Boxplot")
plt.show()

plt.imshow(df[num_cols].corr())
plt.colorbar()
plt.title("Correlation heatmap")
plt.show()

grouped = df.groupby(cat_cols[0])[num_cols[0]].agg(
    ["mean", "median", "count", "sum"]
)
print(grouped.head())

pivot = pd.pivot_table(
    df,
    values=num_cols[0],
    index=cat_cols[0],
    aggfunc=["mean", "count"]
)
print(pivot.head())
```
